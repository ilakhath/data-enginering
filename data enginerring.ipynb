```python
df_col_fil=df[["Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell System Frame Number","Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell Slot Number"]]
df_col_fil
```

```python
#removing null value columns
df_filter=df_col_fil.dropna(subset=["Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell System Frame Number",
                                    "Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell Slot Number"])
```

```python
#current_line column
df_filter['current_line']=(df_filter['Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell System Frame Number']*10)+(df_filter['Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell Slot Number']*0.5)
```

```python
df_filter.shape

```

```python
#previous_line column
df_filter['previous_line']=(df_filter['Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell System Frame Number'].shift(1)*10)+(df_filter['Qualcomm 5G-NR MAC PDSCH Status[Per Slot] PCell Slot Number'].shift(1)*0.5)
```

```python
df_filter.head()
```

```python
#delta_calculation
df_filter['delta']=(df_filter['current_line']-df_filter['previous_line'])/0.5
```

```python
df_filter.head()
```

```python
#if current line lessthan
df_filter['5G-NR_Delta_Pcell (RX_on)']=np.where(df_filter.current_line< df_filter.previous_line,((1023*10+19*0.5)-(df_filter.previous_line)) + (df_filter.current_line) / 0.5,((1023*10+9*1)-(df_filter.previous_line)) + (df_filter.current_line) / 1.0)
```

```python
#if current line greaterthan
df_filter['5G-NR_Delta_Pcell (RX_off)']=np.where(df_filter.current_line> df_filter.previous_line,((1023*10+19*0.5)-(df_filter.previous_line)) + (df_filter.current_line) / 0.5,((1023*10+9*1)-(df_filter.previous_line)) + (df_filter.current_line) / 1.0)
```
