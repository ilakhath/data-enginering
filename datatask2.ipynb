```python
df_fil['delta']=(df_fil['current_line']-df_fil['previous_line'])/0.5
```

```python
#delta column
#if current lin lessthan
df_fil['5G-NR_Delta_Scell (RX_on)']=np.where(df_fil.current_line< df_fil.previous_line,((1023*10+19*0.5)-(df_fil.previous_line)) + (df_fil.current_line) / 0.5,((1023*10+9*1)-(df_fil.previous_line)) + (df_fil.current_line) / 1.0)
```
